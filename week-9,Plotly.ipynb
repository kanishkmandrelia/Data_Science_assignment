#### Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot:
- Distplot is a function from the Seaborn library, primarily used to visualize the distribution of a dataset.
- It combines multiple plots, such as histograms, kernel density estimation (KDE), and optionally a rug plot. 

```python
df4 = px.data.iris()

dist_plot = px.histogram(df4, x="sepal_length", nbins=30, histnorm="density", title="Distribution of Sepal Length", labels={"sepal_length": "Sepal Length"}, template="plotly_white")
# Show the plot
dist_plot.show()
```
