### Q1. What is a database? Differentiate between SQL and NoSQL databases.

Ans.

A database is the collection of data in a structed way. It is desifned and used to store, retrieve and manage data. Used to handle large of amount of data, ensuring easy accessiblility, security and scalability.  

Difference between SQL and NoSQL databases:  

SQL=>  
- Data model - Relational (tables with rows and columns).  
- Schema - Structed schema with predefined tables and data types.  
- Query language - Structure Query language (SQL).  
- Examples - MySQL, PostgreSQL, Oracle, etc.  
- Best for - Complex queries, structured data.  
- Performance - Generally slower for large scale, unstructed data.  

NoSQL=>  
- Data model - Non - Relational (document, key-value, column).  
- Schema - Dynamic schema (allows flexible data structure).  
- Query language - Typically uses queries in formats like JSON or other key-value pair.  
- Examples - MongoDB, Cassandra, Couchbase, etc.  
- Best for - Unstructured or semi structured data, large sacle data, real time application.  
- Performance - Optimised for large datasets.

### Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

Ans.

DDL:
- Data Definition Language.
- Set of SQL commands used to define, modify, and manage the structe of the database, including its schema and table.  
- Deals with the creation, alteration and deletion of database structures like tables, indexes, etc.  
- DDL commands - CREATE, ALTER, DROP, TRUNCATE.  

1. CREATE:  
- Used to create a new database object.  
- Example: Creating a table called employees  
  
  

2. DROP:
- Used to delete an existing database object, such as a table or database, permanently from the system. This operation is irreversible, and any associated data is lost.  
- Example: Droping table name called employees  
  
  

3. ALTER:  
- Used to modify the structure of an existing database object, such as adding, modifying, or dropping columns from a table.  
Example: Adding new column "phone_number" to the "employees" table and modigying the "email" to increase its size  
  
  

4. TRUNCATE:  
- Used to remove all records from a table but keep the table structure for future use. Unlike DROP, TRUNCATE doesnt remove the table but clears its data quickly and efficiently.
Example: Truncating the "employees" table

### Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

Ans.

DML:
- Data Manipulation Language.  
- Used to manipulate data stored in database objects.  
- DML commands - INSERT, UPDATE, DELETE.  

1. INSERT:  
- Used to add new records (rows) into the table.  
- Example: inserting new employee record into "employees" table  
  
  

2. UPDATE:  
- Used to modify existing records in a table.  
- Example: Updating the "email id" of the employee with employee_id = 101  
  
  

3. DELETE:  
- Used to remove one or more rows or all rows at once from table.  
- Example: delteing the employee with employee_id = 101

### Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.

Ans.

```python
import mysql.connector

# Step 1: Establish a connection to the MySQL database
connection = mysql.connector.connect(
    host="localhost",         # Your MySQL host
    user="root",      # Your MySQL username
    password="Kushal@123",  # Your MySQL password
    database="sys"   # The database you want to connect to
)

# Step 2: Create a cursor object
cursor = connection.cursor()

# Step 3: Write an SQL query and execute it
query = "SELECT * FROM employees"
cursor.execute(query)

# Step 4: Fetch and print the results
results = cursor.fetchall()  # Fetch all rows from the executed query

for row in results:
    print(row)

# Step 5: Close the cursor and connection
cursor.close()
connection.close()
```

1. cursor() Method:

The cursor() method creates a cursor object that allows to interact with the database. This object is used to execute SQL queries and fetch results.  

2. execute() Method:

The execute() method is used to run SQL queries on the database through the cursor. The method can execute any type of SQL query, such as SELECT, INSERT, UPDATE, or DELETE.
