#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans.

Matplotlib is a popular data visualization library in Python that provides tools to create static, animated, and interactive visualizations. It allows users to generate a variety of plots and customize them extensively, making it an essential tool for data analysis and exploration.  

Matplotlib is used for:  

1. Data Visualization: To represent data graphically for better understanding and insights.  
2. Customization: To create highly customizable plots tailored to specific needs.  
3. Publication-Ready Graphics: To generate high-quality plots suitable for academic papers or presentations.  
4. Compatibility: It works well with libraries like NumPy and Pandas, enabling seamless integration for data processing and visualization.  

Five Plots Using Pyplot Module of Matplotlib:  
1. Line Plot (plt.plot)-  
Used to represent trends or continuous data over time.  
Example: Plotting temperature changes over a week.  

2. Bar Plot (plt.bar)-  
Used to display categorical data with rectangular bars.  
Example: Comparing sales of different products.  

3. Scatter Plot (plt.scatter)-  
Used to visualize relationships or correlations between two variables.  
Example: Showing the relationship between study hours and test scores.  

4. Histogram (plt.hist)-  
Used to show the frequency distribution of a dataset.  
Example: Analyzing the distribution of students' grades.  

5. Pie Chart (plt.pie)-  
Used to represent proportions or percentages of categories within a whole.  
Example: Displaying market share of different companies.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Note: Also add title, xlabel, and ylabel to the plot.

Ans.

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = x, y = y, )
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])  
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])  
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])  
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans.

- The **subplot()** function in Matplotlib is used to create multiple plots within a single figure.  
- It allows dividing the figures into a grid and plotting individual subplots in the grid's specified locations.  
- This is particularly useful for comparing different datasets or visualizations in a structured layout.

```python
x1,y1 = np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])
x2,y2 = np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])
x3,y3 = np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])
x4,y4 = np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.plot(x1, y1, color="blue", marker="o", linestyle="-")
plt.title("Line 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2, 2, 2)
plt.plot(x2, y2, color="blue", marker="o", linestyle="-")
plt.title("Line 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2, 2, 3)
plt.plot(x3, y3, color="blue", marker="o", linestyle="-")
plt.title("Line 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2, 2, 4)
plt.plot(x4, y4, color="blue", marker="o", linestyle="-")
plt.title("Line 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.tight_layout()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Ans.

A bar plot is a graphical representation of categorical data using rectangular bars. The height (or length) of each bar corresponds to the value it represents. Bars can be displayed either vertically or horizontally.  

A bar plot is used to:

- Compare Categories: It makes it easy to compare values across different categories or groups.
  - Example: Comparing sales figures of various products.
- Highlight Differences: It effectively shows disparities in data by using distinct bar lengths.
- Visualize Trends Over Groups: It can be used to display trends or variations across a range of categories.
   - Example: Population in different cities or exam scores for various subjects.
- Simplify Data Interpretation: The visual format is straightforward, making it accessible to non-technical audiences.

```python
plt.bar(x = company, height= profit)
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Ans.

- A box plot (also called a whisker plot) is a graphical representation of the distribution of a dataset. It displays key summary statistics, including the median, quartiles (Q1, Q2, Q3)), and potential outliers.  
- The plot is characterized by a rectangular "box" with "whiskers" extending from either end.  

A box plot is used for:  
1. Visualizing Data Distribution:
   - It shows how data is spread across its range.
   - Example: Examining the distribution of test scores for a class.

2. Identifying Outliers:  
   - It highlights values that are significantly higher or lower than the rest of the data.

4. Comparing Groups:  
   - It facilitates comparison between multiple datasets side by side.
   - Example: Comparing monthly sales data for different regions.

5. Summarizing Data:  
   - It succinctly conveys key statistics such as the median, quartiles, and range without displaying the raw data.

```python
plt.boxplot(x = [box1, box2])
```
