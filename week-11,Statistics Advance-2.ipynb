#### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

Ans.

1.Probability Mass Function (PMF):  
- PMF is used for discrete random variables.
- It gives the probability of a discrete random variable taking on a specific value.
- The sum of all probabilities in the PMF equals 1.  

2.Probability Density Function (PDF):  
- PDF is used for continuous random variables.
- It describes the likelihood of a random variable taking on a value within a given range.
- Unlike PMF, the value of the PDF at a specific point is not a probability. The area under the curve of the PDF over an interval represents the probability of the variable falling within that interval.

**Example:**  
- PMF Example
Consider a dice roll. The possible outcomes are 1,2,3,4,5,6. The PMF of a fair dice roll assigns a probability of 1/6 to each outcome.

- PDF Example
Consider a continuous random variable X that follows a normal distribution with mean μ=0 and standard deviation σ=1. The PDF of this distribution describes the likelihood of X being within a certain range.

#### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Ans.

**Cumulative Density Function (CDF):**  
The Cumulative Density Function (CDF) gives the cumulative probability of a random variable X being less than or equal to a specific value x:
F(x) = P(X <= x>)
- For discrete random variables, the CDF is the sum of the probabilities up to x.
- For continuous random variables, the CDF is the integral of the PDF from −∞ to x.

**Use of CDF:**  
- Probabilities for Intervals: It allows computation of the probability of a random variable lying within a range.    
- Understanding Distribution: It provides a cumulative perspective on the distribution of the random variable.  
- Simulation & Inverse Transform Sampling: The CDF is used in generating random variables from a given distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Discrete Example: CDF of Dice Roll
dice_outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6  # Fair dice probabilities
cdf_discrete = np.cumsum(probabilities)

plt.step(dice_outcomes, cdf_discrete, where='post', color='blue', label="CDF")
plt.scatter(dice_outcomes, cdf_discrete, color='red', zorder=5)
plt.title("CDF of a Fair Dice Roll")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.legend()
plt.show()

# Continuous Example: CDF of Normal Distribution
x = np.linspace(-4, 4, 500)
mean, std_dev = 0, 1
cdf_continuous = norm.cdf(x, mean, std_dev)

plt.plot(x, cdf_continuous, color='blue', label='CDF')
plt.title("CDF of a Normal Distribution")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.legend()
plt.show()
```

#### Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

Ans.

The normal distribution is commonly used to model real-world phenomena due to its natural occurrence in various domains. Examples include:  
1.Heights of People:
- Heights of adults of the same gender and similar age groups often follow a normal distribution.
- The mean (μ) represents the average height, while the standard deviation (σ) reflects the variability.

2.Test Scores:  
- Standardized test scores (e.g., SAT, GRE) are often modeled as normal distributions to measure performance.
- The mean represents the average score, and the standard deviation measures the spread of scores.

3.Measurement Errors:
- Errors in scientific measurements (e.g., weights, lengths, and temperatures) often follow a normal distribution when the errors are random and independent.

4.Stock Market Returns:  
- Daily returns of stocks or financial instruments are often assumed to be normally distributed in short-term financial models.

5.Blood Pressure:
- Blood pressure measurements in a population tend to follow a normal distribution, where 'μ' represents the average blood pressure and 'σ' represents the variability.

**Parameters of the Normal Distribution and Their Impact:**
1.Mean (μ):
- The mean determines the center or location of the peak of the distribution.
- Shifting μ moves the entire curve left or right along the x-axis without affecting its shape.

2.Standard Deviation (σ):
- The standard deviation controls the spread or width of the distribution.
- A smaller σ produces a narrower and taller curve (values are concentrated near the mean).
- A larger σ produces a wider and flatter curve (values are more spread out).

#### Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

Ans.

**Importance of Normal Distribution:**  
The normal distribution is one of the most important and widely used probability distributions in statistics and data analysis. Its significance lies in several key factors:  
1.Central Limit Theorem (CLT):  
- The CLT states that the sampling distribution of the sample mean will approach a normal distribution, regardless of the original distribution of the data, as the sample size becomes large.
- This makes the normal distribution essential for inferential statistics, hypothesis testing, and confidence intervals.

2.Natural Occurrence:  
- Many real-world phenomena naturally follow a normal distribution due to the cumulative effect of multiple small, independent factors.

3.Simplicity and Applicability:  
- The normal distribution is mathematically convenient, with well-defined properties like symmetry and a single peak at the mean.
- Its parameters (μ and σ) make it easy to model, compare, and analyze datasets.

4.Basis for Statistical Models:  
- Many statistical methods and machine learning algorithms assume that data or errors are normally distributed.
- Common tests like t-tests, ANOVA, and regression analysis rely on this assumption.

**Real-Life Examples of Normal Distribution:**  
1.Human Characteristics:  
- Heights and Weights: Heights of people in a population often follow a normal distribution, centered around the average height with some variation.
- IQ Scores: Intelligence quotient (IQ) scores are typically modeled using a normal distribution with a mean of 100 and a standard deviation of 15.

2.Measurement Errors:  
- Random errors in scientific experiments often follow a normal distribution, assuming the errors are independent and additive.

3.Finance and Economics:  
- Stock Returns: Daily price changes or returns of stocks are often assumed to be normally distributed in financial modeling.
- Income Distribution: While raw income distributions are often skewed, transformations or logarithmic income distributions tend to approximate a normal distribution.

4.Natural Phenomena:  
- Temperature Variations: Daily temperatures in a specific location can often be modeled as normally distributed.
- Test Scores: Scores on standardized tests like the SAT or GRE are designed to approximate a normal distribution to facilitate comparisons.

5.Quality Control in Manufacturing:  
- Measurements of product dimensions (e.g., length, weight, or thickness) tend to follow a normal distribution if the production process is stable.

#### Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

Ans.

```python
from scipy.stats import norm

m = 50  # mean
std = 10 # standard deviation
x = 60

z = (x - m) / std

p_greater_than_60 = 1 - norm.cdf(z)

print(f"Probability of X > 60: {p_greater_than_60:.4f}")
```

#### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans.

**Central Limit Theorem (CLT):**  
- The Central Limit Theorem (CLT) is one of the most important concepts in statistics. It states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables will approximate a normal distribution (Gaussian distribution), regardless of the shape of the original distribution of the data. This is true if the sample size is sufficiently large.

**Significance of the Central Limit Theorem:**  
1.Foundation for Inferential Statistics:  
- The CLT is the backbone of many statistical methods, including hypothesis testing, confidence intervals, and regression analysis. It allows statisticians to make inferences about a population using the sample mean and standard deviation, even when the underlying population distribution is unknown.  

2.Sampling Distributions:  
- CLT enables us to use the sampling distribution of the sample mean to estimate population parameters (mean, standard deviation) and calculate probabilities associated with sample statistics.

3.Approximation to Normal Distribution:  
- Even when the original data is not normally distributed, CLT ensures that with large enough sample sizes, the distribution of the sample means will tend to follow a normal distribution, making it easier to apply statistical techniques that assume normality.

4.Generalizing Results:  
- The theorem provides a powerful way to generalize results from a sample to the population. It allows statisticians to assume normality for the sampling distribution of the sample mean, which facilitates the use of parametric tests even with non-normal data, provided the sample size is large enough.

#### Q10: State the assumptions of the Central Limit Theorem.

Ans.

**Assumptions of the Central Limit Theorem:**  
1.Independence of Observations:  
- The data points (or observations) must be independent of each other. This means that the outcome of one observation should not affect the outcome of another. For example, in a sampling scenario, one sample should not influence the next sample.  

2.Identically Distributed:  
- The observations must be identically distributed, meaning that each data point is drawn from the same probability distribution with the same parameters (mean and variance). This ensures consistency in the data’s characteristics across the sample.  

3.Random Sampling: 
- The data points should be randomly selected from the population. Random sampling ensures that each observation has an equal chance of being chosen, making the sample representative of the population.  

4.Sample Size (n) is Large Enough:  
- The sample size n needs to be sufficiently large for the CLT to hold. While there is no strict rule for what qualifies as "large enough," a general guideline is:  
  - For populations with a normal distribution, even small sample sizes (e.g., n >= 30) are often sufficient.  
  - For populations with non-normal distributions (e.g., skewed or bimodal), larger sample sizes (e.g., n >= 50 or more) are recommended to better approximate a normal distribution.  

5.Finite Variance:  
- The population from which the sample is drawn must have a finite variance (i.e., the data should not have infinite variance). This is important because infinite variance can lead to issues when trying to calculate the standard error or applying the normal distribution approximation.  

6.Population Distribution:  
- The shape of the population distribution doesn't need to be normal. The power of the CLT is that regardless of the population distribution, the distribution of the sample mean will approximate normality as the sample size increases. However, the more skewed the original population distribution is, the larger the sample size needed to reach a good approximation of normality.
