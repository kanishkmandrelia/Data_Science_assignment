Q1. What is Exception in Python? Write the difference between Exceptions and syntax errors.

Ans.

An exception in Python is an event that occurs during the execution of a program that disrupts the normal flow of the program's instructions. It is a Python object that represents an error. When an exception occurs, the Python interpreter stops executing the current code and looks for a way to handle the exception.  
Here are some examples of exceptions in Python:  
1. ZeroDivisionError: This exception is raised when you try to divide a number by zero.
2. KeyError: This exception is raised when you try to access a key that does not exist in a dictionary.
3. NameError: This exception is raised when you try to use a variable that has not been defined.
4. TypeError: This exception is raised when you try to perform an operation on an object that is of the wrong type.

A syntax error is a mistake in the syntax of a Python program. It is an error that occurs when the Python interpreter cannot understand the code. Syntax errors are usually caused by typing mistakes, missing characters, or incorrect indentation.

The main difference between an exception and a syntax error is that an exception is raised during the execution of a program, while a syntax error is detected by the Python interpreter before the program even starts running.

Q2. What happens when an exception is not handled? Explain with an example

Ans.

If an exception is not handled, the Python interpreter will stop executing the current code and print an error message. The error message will usually include the type of exception that was raised, as well as the line number where the exception occurred.  
Example-

```python
# Zero division error
a=10
b=0
print(a/b)
print("Hello world!!")
```

In this cases, it is better to handle exceptions than to let them go unhandled. This is because handling exceptions allows to gracefully handle errors and prevent programs from crashing.

```python
a=10 
b=0
try:
    print(a/b)
except ZeroDivisionError as e:
    print(e,": Can not be divided by 0")
print("Hello World!!")
```

Q3. Which Python statements are used to catch and handle Exception? Explain with an example

Ans.

The 'try' and 'except' statements are used to catch and handle exceptions in Python. The 'try' statement tells Python to execute a block of code, and the 'except' statement tells Python what to do if an exception is raised in the 'try' block.  
The 'finally' statement is executed regardless of whether or not an exception is raised in the 'try' block. The 'finally' block is often used to close files or release resources. 

```python
try:
    print(100/0)
except ZeroDivisionError as e:
    print(e,": Can not be divided by 0")
finally:
    print("THis statement will be executed always")
```

Q4. Explain with an example:  
a. try and else.  
b. finally.  
c. raise.

Ans.

a. Try and else:  
The 'try' and 'else' statements in Python are used to execute a block of code and then execute another block of code if no exceptions are raised in the 'try' block.  
Example:

```python
try:
    print(100/4)
except ZeroDivisionError as e:
    print(e)
else: 
    print("No Exception")
```

b. finally  
The 'finally' statement in Python is used to execute a block of code regardless of whether or not an 'exception' is raised in the 'try' block. The 'finally' block is often used to close files or release resources.  
Example:

```python
try:
    with open("myfile.txt","r") as f:
        read_data = f.read()
except FileNotFoundError as e:
    print(e)
finally:
    print("Thankyou")
    f.close()
```

c. raise  
The 'raise' statement in Python is used to manually raise an 'exception'. The 'raise' statement takes two arguments: the 'exception type' and the 'exception message'.  
Example:

```python
def get_input():
    try:
        age= int(input("enter your age:\t"))
    except ValueError as e:
        raise e

try:
    age= get_input()
except ValueError as e:
    print("Invalid Age")
```

Q5. What is Custom Exception in python? Why do we need Custom Exception? Expalin with example?

Ans.

A Custom Exception is an exception that we define ourself. Custom Exceptions are useful for handling specific errors that are not handled by the built-in exceptions.  
To define a custom exception, we need to create a new class that inherits from the Exception class. The Exception class is the base class for all exceptions in Python.

```python
def validate_age(age):
    if age <= 0:
        raise Exception("Entered age is negative")
    elif age > 150:
        raise Exception("Entered age is very high")
    else:
        print("Age is valid")
        
try:
    age= int(input("enter your age:"))
    validate_age(age)
except Exception as e:
    print(e)
```
