Q1. How do you comment code in Python? What are the different types of comments?

Ans.

Comments in Python are used to explain the code to other programmers, including ourself.
There are two types of comments in Python: 

i. Single-line comments:  These are comments that start with a hash mark (#) and continue to the end of the line.  
Eg:

```python
# Example of Single-line comment
print("Hello, world!")
```

ii. Multi-line comments: These are comments that can span multiple lines. They are created by using three quotation marks (""") to start and end the comment.  
Eg:

```python
"""
Example of  Multi-line comment
that can span multiple lines
"""
print("Hello, world!\nThankyou")
```

Q6. What is a complex number in mathematics, and how is it represented in Python?

Ans.

A complex number is a number in the form of 'a + bi', where 'a' and 'b' are real numbers and 'i' is the imaginary unit, which is defined as the square root of '-1'.  
Complex numbers can be added, subtracted, multiplied, and divided in a similar way to real numbers.  
Representation of complex number in python.

```python
# Normal/ function
print(complex(2,3))

# In list/tuple
l= [1, 6, 2+4j, 5+1j, 4j]
print(l)
```
