Q3. Explain the following operators using an example:

(i)/

(ii)%

(iii)//

(iv)**

Ans.

(i) / =>  This operator is the division operator. It divides the first operand by the second operand and returns the result.

Example

```python
a=10
b=5
div= a/b
div
```

(ii)% => This operator is the modulus operator. It returns the remainder of the division of the first operand by the second operand.

Example

```python
a=19
b=4
rem= a%b
rem
```

(iii)// => This operator is the  floor division operator. It divides the first operand by the second operand and returns the result as an integer. The remainder of the division is discarded.

Example

```python
a=19
b=4
flr= a//b
flr
```

(iv)** => This operator is the exponentiation operator. It raises the first operand to the power of the second operand.

Example

```python
a=3
b=2
exp= a**b
exp
```

Q5. Using a while loop, verify if the number A is purely divisible by number B and if so then how
many times it can be divisible.

Ans.

```python
while (1) :
    A=int(input("value of A:\t"))
    B=int(input("Value of B:\t"))
    print("Value after Division:",A/B)
    while A%B==0:
        print("It is divisible by",A//B,"times")
        break
    if A%B!=0:
        print(A,"is not purely divisible by",B)
        break
```

Q6. Create a list containing 25 int type data. Using for loop and if-else condition print if the element is divisible by 3 or not.

Ans.

```python
lst = range(1,26)
for i in lst:
    print(i)
    if i%3==0:
        print("Yes, ",i,"is divisible by 3:\t",i/3)
        print("\n")
    else :
        print("No, ",i,"is not divisible by 3:\t",i/3)
        print("\n")
```

Q7. What do you understand about mutable and immutable data types? Give examples for both showing this property.

Ans.

Mutable data types can be changed after they are created, while immutable data types cannot be changed.

Example of mutable data types:

1.List

Examples of immutable data types:

1.String

2.Tuple

```python
# Mutable data type
list = [1, 2, 3]
list[0] = 4
print(list)

# Immutable data type
tuple = (1, 2, 3)
tuple[0] = 4
print(tuple)
```
