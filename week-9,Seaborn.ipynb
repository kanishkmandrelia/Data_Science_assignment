#### Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans.

1. Scatter Plot
- Command: sn.scatterplot()
- Use:
  - Displays the relationship between two continuous variables.
  - Helps identify correlations, trends, or outliers in the data.
  - Can include additional information using color (hue), size (size), or style (style).

2. Bar Plot
- Command: sn.barplot()
- Use:
  - Visualizes the distribution of a categorical variable with respect to a numerical variable.
  - Aggregates data (e.g., showing mean or median) and displays confidence intervals.
  - Useful for comparing group means.

3. Box Plot
- Command: sn.boxplot()
- Use:
  - Summarizes the distribution of a numerical variable with key statistics (e.g., median, quartiles).
  - Highlights potential outliers and the spread of the data.
  - Useful for comparing distributions across multiple categories.

4. Heatmap
- Command: sn.heatmap()
- Use:
  - Displays a matrix-like data visualization with color-coded values.
  - Commonly used for correlation matrices or representing tabular data visually.
  - Helps identify patterns, clusters, or correlations.

5. Pair Plot
- Command: sn.pairplot()
- Use:
  - Creates pairwise scatter plots for all combinations of numerical variables in a dataset.
  - Useful for exploratory data analysis (EDA) to examine relationships between variables.
  - Can include distributions along the diagonal and groupings using hue.

#### Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

Ans.

```python
from seaborn import load_dataset
```

```python
df = load_dataset("fmri")
df
```

```python
sn.lineplot(data = df, x = "timepoint", y = "signal", hue= "region")
```

```python
sn.lineplot(data = df, x = "timepoint", y = "signal", hue= "event")
```

#### Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

Ans.

```python
df2 = load_dataset("titanic")
df2
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))

plt.subplot(1,2,1)
sn.boxplot(data= df2, x = "pclass", y = "age")
plt.title("pclass by age")

plt.subplot(1,2,2)
sn.boxplot(data= df2, x = "pclass", y = "fare")
plt.title("pclass by fare")
```
